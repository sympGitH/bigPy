#### 연습
NC 다이노스 2013-2015 시즌에서 다음을 수행
1. 시즌별 홈런 평균
2. 선수별 홈런 평균
3. 시즌별 홈런 평균 이상 홈런을 기록한 선수들의 기록을 홈런 개수 기준 상위 5명 출력

```python
홈런13=프레임목록[0]['홈런']
홈런14=프레임목록[1]['홈런']
홈런15=프레임목록[2]['홈런']
홈런13.name = 2013
홈런14.name = 2014
홈런15.name = 2015

시즌홈런 = pd.concat([홈런13, 홈런14, 홈런15], axis=1)
```

```python
시즌평균홈런 = 시즌홈런.mean()
시즌평균홈런
```

```python
시즌홈런.mean(axis=1)
```

```python
시즌홈런[:3]
```

```python
평균이상홈런=시즌홈런[시즌홈런>=시즌평균홈런]
```

```python
평균이상홈런=평균이상홈런.dropna(thresh=1)
# 한 번 이상이라도 평균 이상 홈런을 친 선수들의 기록
```

```python
홈런합계=평균이상홈런.sum(axis=1).sort_values(ascending=False)
# .rank(ascending=False, method='min')
```

```python
평균이상홈런.ix[홈런합계.index][:5]
```

```python
홈런합계.index
```

```python
def 주석줄찾기(파일명, 시작패턴, encoding='utf-8'):
    주석줄번호목록 = []
    #텍스트 파일 열기
    파일=open(파일명, encoding=encoding)
    for 줄번호, 줄 in enumerate(파일):
        #현재 줄이 시작패턴에 부합하는지 확인
        if 줄.startswith(시작패턴):
            주석줄번호목록.append(줄번호)
    return 주석줄번호목록
```

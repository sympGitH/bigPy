### 누락데이터 처리

```python
프레임목록=importNC('./data/')
```

```python
안타목록 = []

for 프레임 in 프레임목록:
    안타시리즈=프레임['안타']
    안타목록.append(안타시리즈)

시즌안타=pd.concat(안타목록, axis=1)
```

```python
# 열이름 변경
시즌안타.columns=[2013, 2014, 2015]
시즌안타
```

모든 시즌에 출전한 선수들의 기록만 선택
=> NaN이 하나라도 포함된 기록은 제외

__thresh=2__ 를 통해 최소 2개이상 값을 가지는 Row를 선택

```python
시즌안타.dropna(thresh=2)
```

```python
DataFrame(시즌안타.fillna('-') + 시즌안타.fillna('-'))[:3]
# 문자는 문자로 더해지고 숫자는 숫자로 더해짐
```

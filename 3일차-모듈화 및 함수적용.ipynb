```python
import pandas as pd
from NC import importNC
```

```python
프레임목록=importNC('./data/')
```

```python
안타13=프레임목록[0]['안타']
안타14=프레임목록[1]['안타']
안타15=프레임목록[2]['안타']

시즌안타=pd.concat([안타13, 안타14, 안타15], axis=1)
시즌안타
```

```python
def 최대최소값구하기(시리즈):
    return 시리즈.max() - 시리즈.min()
```

```python
시즌안타.apply(최대최소값구하기)
# apply : 함수를 인자로 받아서 적용하는 함수
```

각 색인별로 함수 적용

```python
시즌안타.apply(최대최소값구하기, axis=1)
```

```python
NC15=프레임목록[-1]
# 목록 제일 뒤에 있는 것을 불러옴
```

```python
타율15=NC15['안타'] / NC15['타석']
타율15
```

```python
def 소수점표현(값):
    값 = '{0:.2f}'.format(값)
    return 값

소수점표현(3.1415)
```

시리즈의 각 값에 함수를 적용

```python
타율15.map(소수점표현)
```

```python
타율13=프레임목록[0]['안타'] / 프레임목록[0]['타석']
타율14=프레임목록[1]['안타'] / 프레임목록[1]['타석']
타율15=프레임목록[2]['안타'] / 프레임목록[2]['타석']

타율13.name=2013
타율14.name=2014
타율15.name=2015

시즌타율=pd.concat([타율13, 타율14, 타율15], axis=1)
시즌타율
```

* .apply : 열 또는 행 단위로 함수 적용
* .applymap : 각각의 값에 대해 함수 적용

```python
시즌타율.applymap(소수점표현)
```

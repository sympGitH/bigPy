```python
# 필요 없는 부분은 잘라냄
국제0803=pd.read_excel(경로, skiprows=3, parse_cols='C:K', header=None)
국제0803
```

중간 빈줄 삭제

```python
# 모두 NaN인 줄 삭제
국제0803=국제0803.dropna(how='all')
```

행을 삭제하면서 색인이 고르지 않게 되었습니다. (삭제된 행의 index 가 사라짐)
-> 색인을 변형된 데이터프레임에 맞게 다시 재설정하는 것을 권장합니다.

```python
국제0803=국제0803.reset_index(drop=True)
#reset_index()를 그냥 호출하면 index가 하나 더 생기는 문제점이 있음. (기존 색인이 열이됨) 'drop=True'를 사용하여 제거
```

```python
표범위={'equity':(0,8),
      'fixed income': (8, 16),
      'fx':(16,24),
      'commidity':(24,29),
      'cds':(29,44)
    }
```

```python
# 주어진 표 범위로 표들을 잘라냄
프레임=국제0803

표사전 = {}

for 표제목, 범위 in 표범위.items():
    시작, 종료 = 범위
    표 = 프레임.iloc[시작:종료]
    표사전[표제목]=표
```

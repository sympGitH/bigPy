```python
data=pd.merge(pd.merge(movie_rating, movie_usr), movies)
```

분석 목표 제안 : 성별에 따른 영화 평점의 평균 비교

```python
mean_ratings = data.pivot_table('평점',
                aggfunc='mean',
                index='제목',
                columns='성별')
mean_ratings[:10]
```

평점이 250건 이상은 되어야 하지 않을까?

```python
제목별분류=data.groupby('제목')
```

분류된 각 그룹의 개수

```python
평점개수 = 제목별분류.size()
평점개수충족여부 = 평점개수 > 250
평점개수충족여부[:10]
```

평점 개수가 충분한 자료들만 선택하기

```python
제목목록=평점개수충족여부.index
제목목록
```

```python
mean_ratings[평점개수충족여부]
```

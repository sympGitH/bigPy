### 정수 색인
기본 정수 색인이 아닌 정수 색인이 부여된 경우 색인 선택의 문제

```python
시리즈=pd.Series([0, 1, 2])
```

```python
시리즈.ix[0]
```

```python
시리즈2 = pd.Series(['하나', '둘', '셋'], index=[1, 2, 3])
```

```python
시리즈2
```

색인이 1이라고 하면, 두 가지 중 갈등상황이 발생할 수 있습니다.
1. 기본 정수 색인 1 즉, 두 번째 원소
2. 직접 부여한 정수 색인 1

```python
시리즈2.ix[1]
```

색인이 부여된 경우에도 위치만을 기준으로 선택하기
-> 기본 색인을 이용한 데이터 추출

```python
시리즈2.iloc[1]
# iloc 을 활용하여 기본 색인을 이용한 데이터 추출
```

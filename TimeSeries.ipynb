## 빈도 변환 

### 하방 선택 (Down Sampling)
원래 데이터보다 작은 집합의 기간

금요일 날씨만 고르기

```python
금요일_서울날씨 = 서울날씨['highestTemp']['2014-08'].resample('W-FRI')
```

```python
금요일_서울날씨
```

### 상향 선택 (Up sampling)
원래 데이터보다 많은 기간

일별 기준으로 리샘플링

```python
금요일_서울날씨.resample('D')
```

이전 값으로 보간

```python
금요일_서울날씨.resample('D', fill_method='ffill')
```

경로가 일정한 패턴을 가지고 있다면 문자열 양식을 사용하시는 것을 권장합니다.

```python
경로='./data/NC Dinos {}.xlsx'
```

경로를 패턴으로 표현하면 여러 개의 파일을 읽어들여 데이터프레임으로 생성하는 것도 반복문으로 자동화할 수 있습니다. 두세 개는 직접 입력하는 거소가 큰 차이가 없지만, 그 이상의 개수는 자동화하는 것이 거의 언제나 바람직합니다.

```python
#NC13=pd.read_excel(경로.format(2013))
#NC14=pd.read_excel(경로.format(2014))
#NC15=pd.read_excel(경로.format(2015))
```

```python
프레임목록 = []
for 연도 in range(2013, 2016):
    프레임=pd.read_excel(경로.format(연도))
    프레임['시즌'] = 연도
    프레임=프레임.set_index('선수명')
    프레임목록.append(프레임)
```

```python
프레임목록[2][:5]
```

```python
안타13=프레임목록[0]['안타']
안타14=프레임목록[1]['안타']
안타15=프레임목록[2]['안타']
```

```python
안타13[:3]
```

2015년 NC 타자들의 타율을 구한다. 타율은 안타홈런 합계를 타석으로 나눈 것이다.

```python
NC15=프레임목록[2][['타석', '안타', '홈런']]
```

```python
#NC15안타홈런=(NC15['안타'] + NC15['홈런']) / NC15['타석']
NC15안타홈런=NC15['안타'].add(NC15['홈런'], fill_value=0)
NC15안타홈런=NC15안타홈런.div(NC15['타석'])
NC15['타율']=NC15안타홈런
```

#### 연습
1. 조영훈 ... 최재원 선수의 기록을 선택
2. 색인을 다시 설정하기 위해서는 __.reset_index()__를 사용
3. 범위 선택은 색인에 대해 수행됩니다.

```python
NC15선수=NC15.set_index('선수명')['조영훈':'최재원']
```

```python
%who
```

```python
NC15선수=NC15.set_index('선수명')
```

```python
NC15선수.ix[['테임즈', '나성범'], ['타석', '안타', '홈런']]
```

.ix[:, 열조건] (모든 색인 범위를 선택. 즉, 열선택)

```python
NC15선수.ix[:, ['안타', '홈런']]
```

```python
NC15선수[['안타', '홈런']]
```

열 범위 선택 시, 색인 범위 선택과 분법적인 모호함이 발생합니다.
따라서, 좀더 명확하게 명시해 주기 위해 __.ix[:, 열범위]__ 를 사용합니다.

```python
 NC15선수.ix[:, '타석':'삼진']
```

#### 연습
홈런을 5개 이상 친 선수들의 기록을 골라서 홈런개수가 많은 순으로 3명의 선수를 선택

```python
NC15선수.ix[NC15선수.홈런 >= 5, '타석':'도루'].sort_values(by='홈런', ascending=False)[:3]
```

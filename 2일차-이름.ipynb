```python
import pandas as pd
import os
```

```python
%matplotlib inline
```

```python
이름=pd.read_excel('이름.xlsx')
```

```python
def 상위1000개추출(그룹):
    그룹_1k = 그룹.sort_values(by='출생수', ascending=False)[:1000]
    return 그룹_1k
```

```python
이름1k=이름.groupby(['연도', '성별']).apply(상위1000개추출)
이름1k[:100]
```

분석목표: 시간이 지날수록 더욱 다양한 이름들이 등장할까?

```python
연도별성별비율합계=이름1k.pivot_table('비율', aggfunc='sum', index='연도', columns='성별')
```

```python
연도별성별비율합계.plot(figsize=(12, 6))
```
